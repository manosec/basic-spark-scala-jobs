```scala
import org.apache.spark.sql.SparkSession
import org.apache.spark.sql.functions.{col, year}
```

```scala
val spark = SparkSession.builder().appName("Partition based on year").master("local[*]").getOrCreate()
```

```scala
val sprk_ctx = spark.sparkContext
```

```scala
val rating_gcs_path = "gs://artifacts_spark/rating.csv"

val rating_df = spark.read.format("csv").option("header", true).option("inferSchema", true).load(rating_gcs_path)
```

```scala
rating_df.show(5)
```

```scala
val formattedDf = rating_df.withColumn("year", year(col("timestamp")))
```

```scala
val trimmedDF = formattedDf.limit(100000)
```

```scala
val output_path = "hdfs:///user/mano/5_year_paritioned"

trimmedDF.write.partitionBy("year").format("parquet").save(output_path)
```
